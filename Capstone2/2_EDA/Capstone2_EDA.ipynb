## Histograms

```python
# Selecting all original features for visualization, excluding 'Entity' and 'Year' which are non-numeric
all_original_features = [feature for feature in df.columns if not feature in ['Entity', 'Year']]

# Plot histograms for all original features
num_features = len(all_original_features)
num_cols = 4  # Number of columns in the grid
num_rows = (num_features // num_cols) + (1 if num_features % num_cols > 0 else 0)  # Calculate number of rows required

plt.figure(figsize=(20, 5 * num_rows))
for idx, feature in enumerate(all_original_features, 1):
    plt.subplot(num_rows, num_cols, idx)
    sns.histplot(df[feature], kde=True, bins=30)
    plt.title(feature)
    plt.xlabel('')
    plt.tight_layout()

# Save the plot as a .png file with 300 dpi resolution
plt.savefig("histograms.png", dpi=300)

plt.show()
```

The majority of variables originating from the same energy source exhibit similar, if not virtually identical, data distributions. Using all of them in subsequent modeling stages post-Exploratory Data Analysis (EDA) could yield circular results. Therefore, only one of these variables will be selected for use in the subsequent steps. 

```python
selected_columns = [
    'Entity', 
    'Year', 
    'Geo Biomass Other - TWh', 
    'Solar Generation - TWh', 
    'Wind Generation - TWh', 
    'Hydro Generation - TWh',
    'Biofuels Production - TWh - Total',
]

df_selected = df[selected_columns]
```

```python
# Selecting all original features for visualization, excluding 'Entity' and 'Year' which are non-numeric
all_original_features = [feature for feature in df_selected .columns if not feature in ['Entity', 'Year']]

# Plot histograms for all original features
num_features = len(all_original_features)
num_cols = 4  # Number of columns in the grid
num_rows = (num_features // num_cols) + (1 if num_features % num_cols > 0 else 0)  # Calculate number of rows required

plt.figure(figsize=(20, 5 * num_rows))
for idx, feature in enumerate(all_original_features, 1):
    plt.subplot(num_rows, num_cols, idx)
    sns.histplot(df_selected[feature], kde=True, bins=30)
    plt.title(feature)
    plt.xlabel('')
    plt.tight_layout()

# Save the plot as a .png file with 300 dpi resolution
plt.savefig("histograms.png", dpi=300)

plt.show()
```

The histograms above show that the variables distributions are mostly skewed, where:

* **Geo biomass** energy is predominantly clustered towards the lower range, interspersed with a few years exhibiting elevated values. 
* **Solar and wind** demonstrated a notable left skew. This suggests an initial substantial contribution that has been gaining momentum in recent years. 
* **Hydro and biofuel** sources displayed a more balanced distribution, signaling a steady energy source throughout the years. 
* **Other renewable sources** exhibited a mild left skew yet maintained a wider distribution, indicating a gradually increasing contribution over time. 


**Summary**: From these histograms, we can infer that there have been shifts in the energy landscape over time, with some renewable sources gaining traction in recent years.

## Heatmap

```python
# Removing 'Year' from the selected columns and recalculating the correlation matrix
selected_columns_without_year = [
    'Geo Biomass Other - TWh', 
    'Solar Generation - TWh', 
    'Wind Generation - TWh', 
    'Hydro Generation - TWh',
    'Biofuels Production - TWh - Total'
]

df_selected_without_year = df[selected_columns_without_year]

# Calculating the correlation matrix for the selected columns without 'Year'
correlation_matrix_without_year = df_selected_without_year.corr()

# Plotting the correlation matrix using a heatmap without 'Year'
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix_without_year, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix of Selected Variables (excluding Year)")
plt.xticks(rotation=45)  
plt.yticks(rotation=0)  
plt.tight_layout()

# Save the plot as a .png file with 300 dpi resolution
plt.savefig("heatmap.png", dpi=300)

plt.show()
```

The heatmap provides a visual representation of the Pearson correlation coefficients between the selected features:

* A strong positive correlation exists between **solar** and **wind** energy generation. The synchronous growth in these two sectors could be attributed to simultaneous advancements or investments in these renewable energy technologies.
* Also, **geo biomass** and **wind** show a positive correlation, although not as strong as the correlation between solar and wind generation. This implies that higher biomass energy generation is associated with higher wind energy generation to some extent. Further analysis might explore if certain factors, like governmental policies or global energy trends, are simultaneously influencing these two energy sectors.
* **Hydro** demonstrates weak to moderate positive correlations with **wind** and **solar**, which might be influenced by factors such as geographical conditions or policies promoting renewable energy sources.
* **Biofuels** does not exhibit strong correlations with the other variables, which suggests that biofuels production might be influenced by different factors compared to the other forms of energy generation, or it might operate independently from them in the context of development, investment, and production.

**Summary**: The heatmap reveals all the correlations among these variables are positive, suggesting that increases in one type of energy generation tend to coincide with increases in others, although to varying degrees. This might be reflective of overall trends in energy production over time, such as a general shift towards or development in renewable energy sources.

## Correlation


```python
# high_corr_pairs variable
high_corr_pairs_temp = correlation_matrix_without_year.stack().reset_index()
high_corr_pairs_temp.columns = ['Feature 1', 'Feature 2', 'Correlation']
high_corr_pairs_temp = high_corr_pairs_temp[high_corr_pairs_temp['Feature 1'] != high_corr_pairs_temp['Feature 2']]
high_corr_pairs_temp = high_corr_pairs_temp[high_corr_pairs_temp['Correlation'].abs() > 0.9]

# Determine grid dimensions
grid_cols = 2  # 2 columns in the grid
num_pairs_subset_temp = len(high_corr_pairs_temp)  # Defining num_pairs_subset_temp
grid_rows = (num_pairs_subset_temp // grid_cols) + (1 if num_pairs_subset_temp % grid_cols > 0 else 0)

# Plot scatter plots in a grid format
plt.figure(figsize=(10, 4 * grid_rows))
for idx, (feature1, feature2, corr) in enumerate(high_corr_pairs_temp.values, 1):
    plt.subplot(grid_rows, grid_cols, idx)
    sns.scatterplot(data=df, x=feature1, y=feature2)
    plt.title(f'{feature1} vs {feature2}\n(Corr: {corr:.2f})')
    plt.tight_layout()

plt.show()
```

**Summary**: The positive relationship and high correlation between solar, wind and biofuels energy reveals that periods (or years) of high solar energy production also tend to have high wind and biofuels energy production.

## Year-on-Year Growth Rates

```python
# Melting the DataFrame to a long format suitable for seaborn lineplot
df_melted = df.melt(id_vars=['Year'], 
                    value_vars=selected_columns_without_year, 
                    var_name='Energy Source', 
                    value_name='TWh')

# Plotting all energy sources on the same graph with different colors
plt.figure(figsize=(15, 6))
sns.lineplot(data=df_melted, x='Year', y='TWh', hue='Energy Source')
plt.title('Energy Production Over Years')
plt.xlabel('Year')
plt.ylabel('TWh')
plt.legend(title='Energy Source', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(True)
# Save the plot as a .png file with 300 dpi resolution
plt.savefig("year.png", dpi=300)
plt.show()
```

The line plots give us a temporal perspective on the evolution of various energy sources:

* **Solar** generation has witnessed a significant uptick, especially in the latter years, indicating a growing adoption and utilization of solar energy. This could be due to advancements in solar technology, decreases in the cost of solar installations, and possibly due to policy initiatives promoting renewable energy sources.

* **Wind** generation has experienced a consistent and significant upward trend. Similar to solar energy, advancements in technology and supportive policies might have contributed to this upward trajectory. The growing global emphasis on renewable energy sources to mitigate climate change could also be a key factor.

* **Geo Biomass** generation shows an overall increase but with noticeable fluctuations. This might suggest that the production is influenced by various factors, such as biomass availability, technological applications, or policy changes that might impact its adoption.

* **Biofuels** show a consistent increase over the years. This might be associated with a rise in the adoption of biofuels as an alternative to fossil fuels, particularly in sectors like transportation. The development of biofuel technology and possibly incentives for cleaner fuel alternatives might drive this growth.

* **Hydro** generation seems to be stable and doesnâ€™t exhibit the same growth trend as wind and solar energy. This could be due to geographical or environmental limitations, as hydroelectric power generation often requires substantial water bodies and specific geographical conditions. Also, existing hydroelectric facilities might have reached their production capacities.

**Summary**: The line plots depict the varied evolution of energy sources, with solar and wind seeing recent sharp rises, hydro and geothermal remaining consistent, and an overall increase in renewable contributions.
