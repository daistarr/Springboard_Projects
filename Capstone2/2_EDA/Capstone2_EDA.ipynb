## Histograms

```python
# Selecting all original features for visualization, excluding 'Entity' and 'Year' which are non-numeric
all_original_features = [feature for feature in df.columns if not feature in ['Entity', 'Year']]

# Plot histograms for all original features
num_features = len(all_original_features)
num_cols = 4  # Number of columns in the grid
num_rows = (num_features // num_cols) + (1 if num_features % num_cols > 0 else 0)  # Calculate number of rows required

plt.figure(figsize=(20, 5 * num_rows))
for idx, feature in enumerate(all_original_features, 1):
    plt.subplot(num_rows, num_cols, idx)
    sns.histplot(df[feature], kde=True, bins=30)
    plt.title(feature)
    plt.xlabel('')
    plt.tight_layout()

plt.show()
```

The histograms above show that the variables distributions are mostly skewed, where:

* **Geo Biomass Other - TWh** has most of its the data points concentrated towards the lower end, with a few years showing higher values.
<br>
* **Solar Generation - TWh, Wind Generation - TWh, Electricity from solar (TWh), Solar Capacity, Geothermal Capacity, Solar (% electricity), Wind (% electricity), Biofuels Production - TWh - Total, Electricity from Wind (TWh) and Wind Capacity** have distributions heavily skewed to the left, indicating minimal solar, wind and geothermal energy contribution in the early years, but has started to pick up in more recent years.
<br>
* **Hydro Generation - TWh, Hydro (% equivalent primary energy), Biofuel Production - Twh - Total** as a more even distribution, indicating it's been a consistent energy source over the years.
<br>
* **Renewables (% equivalent primary energy) and Renewables (% electricity)** show a slight left skew but have a broader distribution, indicating increasing contribution over time. 
<br>

**Summary**: From these histograms, we can infer that there have been shifts in the energy landscape over time, with some renewable sources gaining traction in recent years.

## Heatmap

```python
# Excludes non-numeric columns and calculate the correlation matrix for all numerical columns
correlation_matrix_all_columns = df.corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(20, 18))
sns.heatmap(correlation_matrix_all_columns, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix of All Variables")
plt.xticks(rotation=90)  
plt.yticks(rotation=0)  
plt.tight_layout()
plt.show()
```

The heatmap provides a visual representation of the Pearson correlation coefficients between the selected features:

* **Solar Generation - TWh and Solar Capacity as well as Wind Generation - TWh and Wind Capacity** have a very high positive correlation of 0.99, which is expected since the capacity to generate solar and wind energy is directly linked to the actual generation.

* **Solar-related features (Electricity from solar (TWh), Solar (% electricity), and Solar (% equivalent primary energy))** are highly correlated with each other. 

* **Wind-related features (Electricity from wind (TWh), Wind (% electricity), and Wind (% equivalent primary energy))** are highly correlated with each other

* **Renewables (% equivalent primary energy) and Renewables (% electricity)** are positively correlated with most other renewable energy sources. This makes sense as they represent the overall contribution of renewables.

* **Hydro Generation - TWh and Electricity from hydro (TWh)** are highly correlated, as they both represent hydroelectric generation.

**Summary**: The heatmap reveals high correlations among solar, wind, and hydro features, with capacity directly influencing generation and distinct clusters for each energy type.

## Year-on-Year Growth Rates

```python
# Selecting all numeric features for visualization, excluding 'Entity' and 'Year'
all_numeric_features = [feature for feature in df.columns if feature not in ['Entity', 'Year']]

# Adjust the subplot layout based on the number of features
num_features = len(all_numeric_features)
num_cols = 4  # Number of columns in the grid
num_rows = (num_features // num_cols) + (1 if num_features % num_cols > 0 else 0)  # Calculate number of rows required

# Plot line graphs for all numeric features against Year
plt.figure(figsize=(20, 4 * num_rows))
for idx, feature in enumerate(all_numeric_features, 1):
    plt.subplot(num_rows, num_cols, idx)
    sns.lineplot(data=df, x='Year', y=feature)
    plt.title(f'{feature} over Years')
    plt.xlabel('Year')
    plt.ylabel('')
    plt.tight_layout()

plt.show()
```

The line plots give us a temporal perspective on the evolution of various energy sources:

* **Geo Biomass Other - TWh** has seen a moderate increase over time.

* **Solar Generation - TWh, Solar Capacity, Wind Generation - TWh, and Wind Capacity** remained negligible for a long time but started to rise sharply in recent years.

* **Electricity from solar (TWh), Solar (% electricity), Electricity from wind (TWh), Biofuels Production - TWh - Total** show that solar energy started making a significant contribution around the year 2000, with a sharp ascent in the most recent years.

* **Hydro Generation - TWh**:  has been relatively stable, with some fluctuations but no clear upward or downward trend.

* **Geothermal Capacity** remains stable with minimal changes over the years.

* **Renewables (% equivalent primary energy) and Renewables (% electricity)** indicate a steady increase in the contribution of renewable energy sources over the years.

* **Other renewables including bioenergy (TWh)** has been relatively stable over the years, with a slight increase recently.

* **Hydro (% electricity)** has been a consistent contributor, with fluctuations but no clear upward or downward trend.

**Summary**: The line plots depict the varied evolution of energy sources, with solar and wind seeing recent sharp rises, hydro and geothermal remaining consistent, and an overall increase in renewable contributions.
